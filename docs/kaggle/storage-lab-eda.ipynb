# Storage Lab: image-format savings & the HEIC tax

A short EDA of the [Image Compression Benchmarks](https://www.kaggle.com/datasets/cleanorlabs/image-conversions-tax) dataset: how much smaller AVIF/WebP are than JPEG at matched perceptual quality (SSIM), and the file-size cost of converting iPhone HEIC to JPG/PNG.

Source & methodology: https://github.com/cleanor-app/cleanor-storage-lab · DOI 10.5281/zenodo.21217372

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi']=120
```

## 1. Bytes to reach the same quality (SSIM 0.95)
AVIF needs far fewer bytes than JPEG for identical perceptual quality.

```python
s=pd.read_csv(f"/kaggle/input/image-conversions-tax/compression-benchmark-summary.csv")
iso=s[(s.metric=="iso_ssim_mean_bytes")&(s.quality_or_target==0.95)].set_index("format")["value"]/1024
ax=iso[["jpeg","webp","avif"]].plot(kind="bar",color=["#94a3b8","#3b93ff","#4f5bff"])
ax.set_ylabel("KB at SSIM 0.95 (lower is better)");ax.set_title("Same quality, fewer bytes");ax.set_xlabel("")
plt.tight_layout();plt.show()
```

## 2. The HEIC tax
Converting an iPhone HEIC to JPG or PNG at matched quality inflates the file for no visual gain.

```python
h=pd.read_csv(f"/kaggle/input/image-conversions-tax/heic-tax-summary.csv")
ax=h.plot(x="heic_q",y=["jpg_tax_x_median","png_tax_x_median"],kind="bar",color=["#ffb020","#ef4444"])
ax.set_ylabel("size multiplier vs HEIC");ax.set_xlabel("HEIC quality");ax.set_title("HEIC to JPG / PNG size tax")
ax.legend(["HEIC to JPG","HEIC to PNG"]);plt.tight_layout();plt.show()
```

## 3. Cloud storage price at the 2 TB tier

```python
c=pd.read_csv(f"/kaggle/input/image-conversions-tax/cloud-price-index.csv")
tb2=c[c.tier_gb==2000].sort_values("usd_per_gb_month")
ax=tb2.plot(x="provider",y="usd_per_gb_month",kind="barh",color="#0ea5e9",legend=False)
ax.set_xlabel("USD per GB / month");ax.set_ylabel("");ax.set_title("Cheapest $/GB at 2 TB");plt.tight_layout();plt.show()
```
